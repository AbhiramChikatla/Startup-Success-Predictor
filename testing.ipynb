```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a dictionary with the model metrics
data = {
    'Model': ['LGBM', 'XGBoost', 'GradientBoosting', 'AdaBoost', 'Random Forest'],
    'Accuracy': [0.90, 0.88, 0.88, 0.87, 0.92],
    'ROC AUC': [0.85, 0.83, 0.82, 0.81, 0.90],
    'Precision-Recall': [0.94, 0.93, 0.93, 0.93, 0.94]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Melt the DataFrame for easier plotting
df_melted = df.melt(id_vars=['Model'], var_name='Metric', value_name='Score')

# Create the plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=df_melted, x='Model', y='Score', hue='Metric')

# Add value labels on the bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=3)

# Customize the plot
plt.title('Model Performance Comparison', fontsize=14, pad=20)
plt.xlabel('Models', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Metrics', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```
